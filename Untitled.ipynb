```python
# movie_id ,keyword,cast,crew,genres , title,'overwise'
df1 = df1[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
df1.dropna(inplace=True)
```

```python
import ast
def convert_(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])

    return l
```

```python
df1['genres'] = df1['genres'].apply(convert_)
```

```python
df1.head(1)
```

```python
df1['keywords'] = df1['keywords'].apply(convert_)
```

```python
df1.head(1)
```

```python
df1['cast'][0]
```

```python
def cast_convert(obj):
    l=[]
    c=0;
    for i in ast.literal_eval(obj):
        if i != 3:
            l.append(i['name'])
            c+=1
        else:break
    return l
```

```python
df1['cast'] =df1['cast'].apply(cast_convert)
```

```python
df1.head(1)
```

```python
df1.drop('crew',axis=1,inplace=True)
```

```python
df1.head(1)
```

```python
df1['cast'] =df1['cast'].apply(lambda x:x[0:3])
```

```python
df1['overview'] =df1['overview'].apply(lambda x:x.split())
```

```python
df1['genres']=df1['genres'].apply(lambda x:[i.replace(" ","")for i in x])
df1['keywords'] =df1['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
df1['cast'] =df1['cast'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
df1.head(1)
```

```python
df1['tags'] =df1['overview']+df1['genres']+df1['keywords']+df1['cast']
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    for i in text.split():
        y=[]
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df2['tags'] =df2['tags'].apply(stem)
```
